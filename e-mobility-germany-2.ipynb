# E-Cars in Germany: New Registrations vs Charing Infrastructure in 2023

To gain a deeper understanding of the current e-car market in Germany, one factor to look at, is charging infrastructure. First, I will have a look at the KBA data for newly registered cars ("Neuzulassungen") in 2023. I discuss the rise of e-cars and what models are most popular. After that, I will dive into the charging infrastructure in Germany and the relation to the new regsitrations. 

First things first, To get an overview of the current situation of e-cars in germany, let's have a look at new registrations.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/kaggle/input/modified-kba/DEC12.xlsx", sheet_name = None)
df_months = df["months"]
```

```python
df_months.dtypes
```

```python
filter_word = "insgesamt"
filtered = df_months['Monate'].str.contains(filter_word, na=False)

df_per_year= df_months[filtered]
df_per_year = df_per_year.reset_index(drop=True)
```

```python
df_per_year = df_per_year.iloc[::-1].reset_index(drop=True)
df_per_year
```

```python
#Plot the Neuzulassungen of new cars over time per Antriebsart

x_values1 = df_per_year["Monate"]
y_values1 = df_per_year["Anzahl insgesammt"]

x_values2 = df_per_year["Monate"]
y_values2 = df_per_year["Alternative Antriebe"]

x_values3 = df_per_year["Monate"]
y_values3 = df_per_year["Elektro total"]

x_values4 = df_per_year["Monate"]
y_values4 = df_per_year["Elektro BEV"]

plt.plot(x_values1, y_values1, label='Anzahl insgesamt', color='blue', marker='o', linestyle='-', linewidth=2)
plt.plot(x_values2, y_values2, label='Alternative Antriebe', color='red', marker='s', linestyle='--', linewidth=2)
plt.plot(x_values3, y_values3, label='Elektro total', color='green', marker='v', linestyle=':', linewidth=2)
plt.plot(x_values4, y_values4, label='Elektro BEV', color='purple', marker='p', linestyle='-.', linewidth=2)

plt.xlabel('Months')
plt.ylabel('Total Number')
plt.title('Total Number per Month')

plt.grid(True, linestyle='--', alpha=0.5)

plt.xticks(rotation=45)

plt.legend()

plt.tight_layout() 
plt.show()
```

Interesting to see, that even though the total amount of new cars registered has gone down in the last 7 years, the amount of new elektric cars has risen. 

Now, to get a better understanding of the current situation, let's have a look at the last two years. Let's plot the amount of new registered cars per month for 2022 and 2023.

```python
df_months.head(26)
```

```python
df_last_two_years = df_months.loc[:25]
df_last_two_years
```

```python
df_last_two_years.loc[:13, 'Monate'] += " 2023"
df_last_two_years.loc[13:, 'Monate'] += " 2022"
df_last_two_years = df_last_two_years.drop(df_last_two_years.index[[0, 13]])
df_last_two_years.reset_index(drop=True)
```

```python
df_last_two_years = df_last_two_years.iloc[::-1].reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt

# Assuming df_last_two_years contains your data
# Calculate percentage change for each series from the start to the end
start_values = df_last_two_years.iloc[0][["Alternative Antriebe", "Elektro total", "Elektro BEV"]]
end_values = df_last_two_years.iloc[-1][["Alternative Antriebe", "Elektro total", "Elektro BEV"]]
percentage_change = ((end_values - start_values) / start_values) * 100

# Plotting
plt.figure(figsize=(10, 5)) 

x_values2 = df_last_two_years["Monate"]
y_values2 = df_last_two_years["Alternative Antriebe"]
plt.plot(x_values2, y_values2, label='Alternative Antriebe', color='red', marker='o', linestyle='--', linewidth=2)

x_values3 = df_last_two_years["Monate"]
y_values3 = df_last_two_years["Elektro total"]
plt.plot(x_values3, y_values3, label='Elektro total', color='green', marker='s', linestyle=':', linewidth=2)

x_values4 = df_last_two_years["Monate"]
y_values4 = df_last_two_years["Elektro BEV"]
plt.plot(x_values4, y_values4, label='Elektro BEV', color='purple', marker='^', linestyle='-.', linewidth=2)

# Add annotation for percentage change slightly above or below the end of each line
for i, month in enumerate(df_last_two_years['Monate']):
    plt.annotate(f'{percentage_change["Alternative Antriebe"]:.2f}%', (x_values2.iloc[-1], y_values2.iloc[-1]), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=9, color='red')
    plt.annotate(f'{percentage_change["Elektro total"]:.2f}%', (x_values3.iloc[-1], y_values3.iloc[-1]), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=9, color='green')
    plt.annotate(f'{percentage_change["Elektro BEV"]:.2f}%', (x_values4.iloc[-1], y_values4.iloc[-1]), textcoords="offset points", xytext=(0, -20), ha='center', fontsize=9, color='purple')

plt.xlabel('Months', fontsize=12, fontweight='bold')
plt.ylabel('Total Number', fontsize=12, fontweight='bold')
plt.title('Total Number per Month', fontsize=14, fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.5)

plt.xticks(rotation=45)
plt.yticks(fontsize=10)

plt.legend()

plt.tight_layout() 
plt.show()
```

Now that we have the plotted values for the last 2 years, we can see that the number of registered electric cars has risen spectacularly in the last two years, with pure electric cars (BEV) a staggering 161.6 percent.

Next, let's try and see how the distribution is per bundesland.

```python
import geopandas as gpd
df_bundesland = pd.read_excel("/kaggle/input/bundesland/KBA-Bundesland.xlsx")
df_bundesland.head(5)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# Load the shapefile for Germany
germany = gpd.read_file("/kaggle/input/de-bund")

# Merge Germany GeoDataFrame with df_bundesland based on the common column 'GEN'
germany = germany.merge(df_bundesland, left_on='GEN', right_on='Bundesland', how='left')

# Calculate the percentage of registrations per Bundesland
germany['Percentage'] = (germany['Anzahl insgesamt'] / germany['Anzahl insgesamt'].sum()) * 100

fig, ax = plt.subplots(figsize=(12, 8))  # Adjusted figure size

germany.plot(column='Anzahl insgesamt', cmap='coolwarm', linewidth=0.8, edgecolor='0.8', legend=False, ax=ax)

# Add labels of Bundeslnder to the map with adjusted positions
for _, row in germany.iterrows():
    if row['Bundesland'] == 'Brandenburg':
        ax.annotate(text=f"{row['Bundesland']}\n{row['Percentage']:.2f}%", xy=(row.geometry.centroid.x, row.geometry.centroid.y + 0.4), xytext=(3, 3), textcoords="offset points", fontsize=8, ha='center', color='black')
    else:
        ax.annotate(text=f"{row['Bundesland']}\n{row['Percentage']:.2f}%", xy=(row.geometry.centroid.x, row.geometry.centroid.y), xytext=(3, 3), textcoords="offset points", fontsize=8, ha='center', color='black')

cbar = plt.colorbar(ax.collections[0], ax=ax, shrink=0.7, aspect=15)  
cbar.set_label('Total Number')

cbar.locator = MaxNLocator(nbins=6)  
cbar.update_ticks()

# Set title
ax.set_title('Number of Registrations E-Pkws per Bundesland 2023', fontsize=20)

plt.tight_layout()

plt.show()
```

Clearly, both Bayern and Nordrhein-Westfalen have the most new registered electric cars in the last year (2023).

# Charging Infrastructure

Now that we have a good idea of the amount of new electric cars in Germany and the distribution per Bundesland, let's have a look at the infrastructure available.

```python
df_charging = pd.read_csv("/kaggle/input/charging-stations/rhein-kreis-neuss-ladesaulen-in-deutschland.csv", delimiter=";")
```

```python
df_charging.head()
```

```python
charging_counts = df_charging["Anzahl Ladepunkte"].groupby(df_charging["Ort"]).count()
charging_counts = charging_counts.reset_index()
charging_counts = charging_counts.sort_values(by="Anzahl Ladepunkte", ascending=False)

charging_counts
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
from matplotlib.colors import Normalize
from mpl_toolkits.axes_grid1 import make_axes_locatable

germany_charge = gpd.read_file("/kaggle/input/cities")

germany_charge = germany_charge.merge(df_charging, left_on='name', right_on='Ort', how='left')

# Plotting
fig, ax = plt.subplots(figsize=(12, 10))  

cmap = plt.cm.get_cmap('Blues', 6)  
germany_charge.plot(column='Anzahl Ladepunkte', cmap=cmap, marker='x', linewidth=0.8, edgecolor='0.8', legend=False, ax=ax)

ax.set_title('Anzahl Ladepunkte in Deutschland', fontsize = 20)

vmin = 1
vmax = 2000

norm = Normalize(vmin=vmin, vmax=vmax)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, fraction=0.03, pad=0.04)
cbar.ax.yaxis.set_major_locator(MaxNLocator(nbins=6))

plt.show()
```

This plot shows the amount of charging stations per city. Let's reduce it to a Bundesland level, to be able to compare it to how many new cars have been registered. 

```python
df_laden_bdl = pd.read_csv("/kaggle/input/laden-pro-bdld/Ladepunkte pro Bundesland - Blad1-2.csv")
```

```python
df_laden_bdl.head()
```

```python
df_laden_bdl['01.10.23'] = df_laden_bdl['01.10.23'].astype(str).str.replace(',', '').str.replace('.', '')
df_laden_bdl['01.10.23'] = df_laden_bdl['01.10.23'].astype(int)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

germany = gpd.read_file("/kaggle/input/de-bund")

germany = germany.merge(df_laden_bdl, left_on='GEN', right_on='Bundesland', how='left')

# Calculate the percentage of registrations per Bundesland
germany['Percentage'] = (germany['01.10.23'] / germany['01.10.23'].sum()) * 100

fig, ax = plt.subplots(figsize=(12, 8))  # Adjusted figure size

germany.plot(column='01.10.23', cmap='coolwarm', linewidth=0.8, edgecolor='0.8', legend=False, ax=ax)

# Add labels of Bundeslnder to the map with adjusted positions
for _, row in germany.iterrows():
    if row['Bundesland'] == 'Brandenburg':
        ax.annotate(text=f"{row['Bundesland']}\n{row['Percentage']:.2f}%", xy=(row.geometry.centroid.x, row.geometry.centroid.y + 0.4), xytext=(3, 3), textcoords="offset points", fontsize=8, ha='center', color='black')
    else:
        ax.annotate(text=f"{row['Bundesland']}\n{row['Percentage']:.2f}%", xy=(row.geometry.centroid.x, row.geometry.centroid.y), xytext=(3, 3), textcoords="offset points", fontsize=8, ha='center', color='black')

cbar = plt.colorbar(ax.collections[0], ax=ax, shrink=0.7, aspect=15)  
cbar.set_label('Total Number')

cbar.locator = MaxNLocator(nbins=6)  
cbar.update_ticks()

# Set title
ax.set_title('Total Number of Charging Stations per Bundesland', fontsize=20)

plt.tight_layout()

plt.show()
```

Now, let's find out how many new e-cars per charging stations there are for each Bundesland.

```python
df_bundesland
```

```python
cars_per_ladepunkt = pd.merge(df_bundesland, df_laden_bdl[['Bundesland', '01.10.23']], on="Bundesland", how="inner")
cars_per_ladepunkt
```

```python
cars_per_ladepunkt["Elektro_total_pro_ladepunkt"] = cars_per_ladepunkt["Elektro Total"]  / cars_per_ladepunkt["01.10.23"]
```

```python
cars_per_ladepunkt_sorted = cars_per_ladepunkt.sort_values("Elektro_total_pro_ladepunkt", ascending=True)
cars_per_ladepunkt_sorted
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  
bars = plt.bar(cars_per_ladepunkt_sorted['Bundesland'], cars_per_ladepunkt_sorted['Elektro_total_pro_ladepunkt'])

plt.xlabel('Bundesland')
plt.ylabel('Number of Cars per Ladepunkt')
plt.title('#Newly Registered E-Cars per Charging Station by Bundesland (2023)', fontsize = 20)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center')

plt.xticks(rotation=45, ha='right', fontsize=10) 

plt.tight_layout()  
plt.show()
```

Initially, it seemed that those Bundeslnder where more e-cars are being bought seemed to have better infrastructure, but that is not the case everywhere and it's not the whole story. In Bremem and Saarland, for example, the ratio of cars vs charging infrastructure is a lot lower then elsewhere.
